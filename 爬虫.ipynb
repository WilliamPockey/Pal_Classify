```python
import requests
from bs4 import BeautifulSoup
import os

def download_images(keyword, max_images):
    url = "https://www.google.com/search?site=&tbm=isch&source=hp&biw=1873&bih=990&q=" + keyword
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
    }

    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')
    images = soup.find_all('img')

    image_count = 0
    for i, image in enumerate(images):
        try:
            image_url = image['src']
            response = requests.get(image_url, headers=headers, stream=True)
            response.raise_for_status()

            with open(f'img_{keyword}_{i}.jpg', 'wb') as fd:
                for chunk in response.iter_content(chunk_size=50000):
                    fd.write(chunk)

            image_count += 1
            if image_count >= max_images:
                break

        except Exception as e:
            pass

    print(f"Downloaded {image_count} images.")

# 使用函数
download_images('绵悠悠', 40)
download_images('爆电熊', 40)
download_images('捣蛋猫', 40)
download_images('翠叶鼠', 40)
download_images('疾旋鼬', 40)
download_images('瞅什魔', 40)
download_images('佩克龙', 40)
download_images('融焰娘', 40)
download_images('雷胖达', 40)
download_images('米露菲', 40)

```

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import requests
import os

def download_images(keyword, max_images):
    url = "https://www.google.com/search?site=&tbm=isch&source=hp&biw=1873&bih=990&q=" + keyword

    # 使用 Selenium 打开网页
    driver = webdriver.Firefox()  # 或者使用其他浏览器，例如：webdriver.Chrome()
    driver.get(url)

    # 滚动页面以加载更多的图片
    for _ in range(5):  # 修改这个值可以控制滚动的次数
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # 使用 BeautifulSoup 解析页面内容
    soup = BeautifulSoup(driver.page_source, 'html.parser')

    # 关闭浏览器
    driver.close()

    # 下载图片
    images = soup.find_all('img')
    image_count = 0
    for i, image in enumerate(images):
        try:
            image_url = image['src']
            response = requests.get(image_url, stream=True)
            response.raise_for_status()

            with open(f'img_{keyword}_{i}.jpg', 'wb') as fd:
                for chunk in response.iter_content(chunk_size=50000):
                    fd.write(chunk)

            image_count += 1
            if image_count >= max_images:
                break

        except Exception as e:
            pass

    print(f"Downloaded {image_count} images.")

# 使用函数
download_images('幻兽帕鲁绵悠悠', 50)
```
